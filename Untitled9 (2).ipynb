```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("IPL2025Batters.csv")

```

```python
df.head()
```

plt.figure() creates a new figure or canvas for your plot.

plt.scatter() draws a scatter plot using points.

plt.bar() draws a vertical bar chart.

plt.barh() draws a horizontal bar chart.

plt.plot() draws a line plot.

plt.hist() draws a histogram

Which 10 batters have the highest strike rate (min 150 runs)?



```python
import matplotlib.pyplot as plt
qualified = df[df["Runs"] >= 150].sort_values("SR", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.bar(qualified["Player Name"], qualified["SR"], color="coral")
plt.xticks(rotation=60, ha="right")
plt.ylabel("Strike Rate")
plt.title("Top 10 Batters by Strike Rate (Min 150 Runs) - IPL 2025")
plt.show()
```

color= changes the color of bars, points, or lines.

label= gives a name to the plotted data, useful for legends.

alpha= controls transparency.

marker= changes the point style in line or scatter plots.

linestyle= changes the line style like solid, dashed, or dotted.

linewidth= changes line thickness.



Which 10 batters scored the most runs in IPL 2025?




```python
top10_runs = df.sort_values("Runs", ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top10_runs["Player Name"], top10_runs["Runs"], color="skyblue")

plt.xlabel("Runs")
plt.title("Top 10 Run Scorers - IPL 2025")

plt.show()
```

How many total runs did each team’s batters score?

```python
team_runs = df.groupby("Team")["Runs"].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(team_runs.index, team_runs.values, color="orange")
plt.xticks(rotation=45, ha="right")
plt.ylabel("Total Runs")
plt.title("Total Runs by Team - IPL 2025")
plt.show()
```

plt.xlabel() sets the x-axis label.

plt.ylabel() sets the y-axis label.

plt.title() sets the chart title.

plt.xticks() changes x-axis tick labels, rotation, and font settings.

plt.yticks() changes y-axis tick labels

Is there any relationship between a batter’s total runs and their strike rate?


```python
plt.figure(figsize=(8, 6))
plt.scatter(df["Runs"], df["SR"], alpha=0.7)
plt.xlabel("Runs")
plt.ylabel("Strike Rate")
plt.title("Runs vs Strike Rate - IPL 2025 Batters")
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

How do players compare based on runs scored per innings?

```python
df["Runs_per_Inn"] = df["Runs"] / df["Inn"]

top15_rpi = df.sort_values("Runs_per_Inn", ascending=False).head(15)

plt.figure(figsize=(10, 6))
plt.bar(top15_rpi["Player Name"], top15_rpi["Runs_per_Inn"], color="purple")
plt.xticks(rotation=60, ha="right")
plt.ylabel("Runs per Innings")
plt.title("Top 15 Batters by Runs per Innings - IPL 2025")
plt.show()
```

 Pie chart of runs share for a team

```python
team_name = "MI"
team_df = df[df["Team"] == team_name]

plt.figure(figsize=(8, 8))
plt.pie(team_df["Runs"], labels=team_df["Player Name"], autopct="%1.1f%%", startangle=140)
plt.title(f"Runs Share Among {team_name} Batters - IPL 2025")
plt.show()
```

plt.grid() adds grid lines to the chart.

Your spelling plt.gird should be plt.grid().

plt.tight_layout() automatically adjusts spacing so labels do not overlap.

plt.legend() shows the legend for labeled plots

Which teams got the most centuries and fifties from their batters?



```python
team_100s_50s = df.groupby("Team")[["100s", "50s"]].sum()

x = range(len(team_100s_50s))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar([i - width/2 for i in x], team_100s_50s["100s"], width=width, label="100s", color="red")
plt.bar([i + width/2 for i in x], team_100s_50s["50s"], width=width, label="50s", color="green")

plt.xticks(list(x), team_100s_50s.index, rotation=45, ha="right")
plt.ylabel("Count")
plt.title("Centuries and Fifties by Team - IPL 2025")
plt.legend()
plt.show()
```

plt.show() displays the final plot on the screen.

plt.subplots() creates a figure and axes together, useful for advanced plotting.



Which team’s batters hit the most boundaries (fours and sixes)?



```python
team_boundaries = df.groupby("Team")[["4s", "6s"]].sum()

plt.figure(figsize=(10, 6))
plt.bar(team_boundaries.index, team_boundaries["4s"], label="4s", color="skyblue")
plt.bar(team_boundaries.index, team_boundaries["6s"],
        bottom=team_boundaries["4s"], label="6s", color="navy")

plt.xticks(rotation=45, ha="right")
plt.ylabel("Total Boundaries")
plt.title("Total 4s and 6s by Team - IPL 2025")
plt.legend()
plt.show()
```
